```python
#rename columns
top_spenders_grouped.rename(columns={'Item ID': 'Purchase Count', 'Price': 'Total Purchase Value'}, inplace=True)
top_spenders_grouped.head()
```

```python
#sort
top_spenders_summary = top_spenders_grouped.sort_values(by='Total Purchase Value', ascending = False)
top_spenders_summary.head()
```

```python
#format amounts
top_spenders_summary['Total Purchase Value']='$'+(top_spenders_summary['Total Purchase Value'].astype(float)).round(2).astype(str)
top_spenders_summary['Average Purchase Price'] = '$' + (top_spenders_summary['Average Purchase Price'].astype(float)).round(2).astype(str)
top_spenders_summary.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#popular items
popular_items.head()
```

```python
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
popular_items_groups=popular_items.groupby(['Item ID', 'Item Name'], as_index=False).agg({'SN':'count', 'Price': 'sum'})
popular_items_groups.head()
```

```python
#rename columns 
popular_items_groups.rename(columns={'SN':'Purchase Count','Price': 'Total Purchase Value'}, inplace=True)
popular_items_groups.head()
```

```python
popular_items_groups['Item Price']=popular_items_groups['Total Purchase Value']/popular_items_groups['Purchase Count']
popular_items_groups.head()
```

```python
#Sort the purchase count column in descending order and format
popular_items_groups2 = popular_items_groups.sort_values(by='Purchase Count',ascending = False)
popular_items_groups2['Total Purchase Value']='$'+(popular_items_groups2['Total Purchase Value'].astype(float)).round(2).astype(str)
popular_items_groups2['Item Price']='$'+(popular_items_groups2['Item Price'].astype(float)).round(2).astype(str)
popular_items_groups2.head()
```
